```python
import pandas as pd
df= pd.read_csv("file.csv")
df.info()
df
```

```python
filtro1 = df[df["Sexo"] == "Mujeres"]
filtro1
```

```python
filtro2  = df[df["Periodo"] == 2022]
filtro2
```

```python
filtro3 = filtro1[df["Entidad_Federativa"] == "Puebla"]
filtro3
```

```python
filtro4 = df.iloc[ [1000, 2000, 3000], :]
filtro4
```

```python
filtro5 = df.iloc[50:550, :]
filtro5
```

```python
filtro6 = df.iloc[ : , 2:5]
filtro6
```

```python
#Para que se cumplan dos filtros se usa el simbolo "&", y para decir este u otro filtro se usa "|"
filtro7 =df[(df["Periodo"] >= 2006) | (df["Sexo"] == "Mujeres")]
filtro7
```
